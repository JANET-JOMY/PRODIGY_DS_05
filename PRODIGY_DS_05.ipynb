
TASK 5:<BR>Analyze traffic accident data to identify patterns related to road conditions, weather, and time of day. Visualize accident hotspots and contributing factors.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#importing data set
df=pd.read_csv("cleaned.csv")
df
df.head(10)
```

```python
df.info()
df.describe()
```

```python
plt.figure(figsize=(8,6))
custom_palette={'Male':'red','Female':'pink','Unknown':'black'}
sns.countplot(x='Sex_of_driver',data=df,palette=custom_palette)
plt.title("Accident count by sex of driver")
plt.show()
```

```python
# Count the occurrences of Weather_conditions and Road_surface_type
weather_counts = df['Weather_conditions'].value_counts()
road_surface_counts = df['Road_surface_type'].value_counts()

# Plot for Weather_conditions
plt.figure(figsize=(10, 6))
weather_counts.plot(kind='bar', color='skyblue')
plt.title('Distribution of Road accidents by Weather Conditions')
plt.xlabel('Weather Condition')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Plot for Road_surface_type
plt.figure(figsize=(10, 6))
road_surface_counts.plot(kind='bar', color='lightgreen')
plt.title('Distribution of Road accidents by Road Surface Types')
plt.xlabel('Road Surface Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Plot for Light_conditions
light_counts = df['Light_conditions'].value_counts()
plt.figure(figsize=(10, 6))
light_counts.plot(kind='bar', color='orange')
plt.title('Distribution of Road accidents by Light Conditions')
plt.xlabel('Light Condition')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(df['Age_band_of_driver'],bins=20,kde=True,color='black')
plt.title("Distribution of Accidents by Age")
plt.xlabel("Age")
plt.ylabel("Number of Accidents")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(df['Type_of_collision'],bins=20,kde=True,color='red')
plt.title("Types of collision")
plt.xlabel("Types of collision")
plt.ylabel("Number of Accidents")
plt.xticks(rotation=30,ha='right')
plt.show()
```

```python
cross_tab = pd.crosstab(df['Educational_level'], df['Accident_severity'])

# Plot the stacked bar chart
plt.figure(figsize=(10, 6))
cross_tab.plot(kind='bar', stacked=True)
plt.title('Educational Level vs. Accident Severity')
plt.xlabel('Educational Level')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend(title='Accident Severity')
plt.tight_layout()
plt.show()
```

```python
cross_tab = pd.crosstab(df['Driving_experience'], df['Accident_severity'])

# Plot the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(cross_tab, cmap="PiYG", annot=True, fmt="d")
plt.title('Driving_experience vs. Accident Severity')
plt.xlabel('Accident Severity')
plt.ylabel('Driving_experience')
plt.tight_layout()
plt.show()
```

```python
cross_tab = pd.crosstab(df['Educational_level'], df['Accident_severity'])

# Plot the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(cross_tab, cmap="YlGnBu", annot=True, fmt="d")
plt.title('Educational Level vs. Accident Severity')
plt.xlabel('Accident Severity')
plt.ylabel('Educational Level')
plt.tight_layout()
plt.show()
```

```python
cross_tab = pd.crosstab(df['Weather_conditions'], df['Road_surface_type'])

# Plot the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(cross_tab, cmap="RdYlGn", annot=True, fmt="d")
plt.title('Weather conditions vs. Road surface types')
plt.xlabel('Weather_Conditions')
plt.ylabel('Road_surface_type')
plt.tight_layout()
plt.show()
```

```python
# Junction types vs Collision types 
junction_collision_radar=df.groupby("Types_of_Junction")['Type_of_collision'].value_counts().unstack().plot(kind='bar',stacked=True)
plt.title("Types of junction vs Types of colllision")
plt.xlabel("Types of  Junction")
plt.ylabel("Types of collision")
plt.show()
```

```python
import geopandas as gpd
from shapely.geometry import Point
import matplotlib.pyplot as plt
stage_mapping = {'Male': 1, "Female": 2, 'Unknown': 3}
df['Sex_of_driver'] = df['Sex_of_driver'].map(stage_mapping)

geometry = [Point(xy) for xy in zip(df['Sex_of_driver'], df['Accident_severity'])]

# Create a GeoDataFrame
gdf = gpd.GeoDataFrame(df, geometry=geometry)

# Load world map data
world = gpd.read_file(gpd.datasets.get_path("naturalearth_cities"))

# Plot the world map
ax = world.plot(figsize=(15, 10), color="blue")

# Plot accident hotspots on top of the world map
gdf.plot(ax=ax, color="red", markersize=10, alpha=0.5)

plt.title("Accident Hotspots - Geospatial Plot")
plt.show()
```

```python
corr_matrix = df[['Sex_of_driver', 'Accident_severity']].corr()

# Plot the correlation heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Heatmap between Sex of Driver and Accident Severity')
plt.show()
```
